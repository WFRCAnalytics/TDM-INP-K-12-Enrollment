```python
import pandas as pd
#import matplotlib.ticker as mtick
import numpy as np
```

```python
df1=pd.read_csv ('results/K12_Enrollment_USTM_TAZ_2021_09_22.csv')
df2=pd.read_csv ('results/privK12_Enrollment_USTM_TAZ_2021_09_22.csv')

# Combine them row-wise (default axis=0)
df_combined = pd.concat([df1, df2], ignore_index=True)
display (df_combined)
```

```python
# Group by CO_TAZID and sum the enrollment columns
df_grouped = df_combined.groupby('CO_TAZID', as_index=False)[['Enrol_Elem', 'Enrol_Midl', 'Enrol_High']].sum()

display(df_grouped)
```

```python
#export CO_TAZID enrollment to csv

df_grouped.to_csv("results\GroupedEnrollmentByTAZ.csv", index=False)
```
